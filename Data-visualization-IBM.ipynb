```python
df = pd.read_csv('Topic_Survey_Assignment.csv',index_col=0)
df.head() 
```

```python
ax=df.plot(kind='bar', figsize=(20, 8), width=0.8, color=['#5cb85c','#5bc0de','#d9534f'])
Topics=["Data Analysis/Statistics","Machine Learning","Data Visualization", "Big Data (Spark/Hadoop)", "Deep Learning", "Data Journalism"]
plt.xticks(fontsize=14)
ax.set_xticklabels(Topics)
ax.set_ylabel("Interest Level in %", fontsize=14)
ax.set_title("Percentage of the respondents interest in the different data science", fontsize = 16)
ax.legend(fontsize=14)
ax.get_yaxis().set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
```
